````python
def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_input = False
        while not valid_input:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity >= 0:
                    inventory[product] = quantity
                    valid_input = True
                else:
                    print("Quantity cannot be negative. Please enter a valid quantity.")
            except ValueError:
                print("Invalid input. Please enter a valid quantity.")
    return inventory
````

````python
def get_customer_orders(inventory):
    valid_input = False
    customer_orders = set()
    while not valid_input:
        try:
            number_order = int(input("Enter the number of customer orders: "))
            if number_order >= 0:
                valid_input = True
            else:
                print("Number cannot be negative. Please enter a valid number of order.")
        except ValueError:
                print("Invalid input. Please enter a number number.")
    
    for i in range(number_order):
        valid_input = False
        while not valid_input:
            try:
                product = input("Enter the name of a product that a customer wants to order: ")
                #print(product, inventory.keys(), inventory[product])
                if product in inventory.keys():
                    if inventory[product] > 0:
                        customer_orders.add(product)
                        valid_input = True
                    else:
                        print("There is no stock left. Please enter a product name")
                else:
                    print("Error with the product name. Please enter a product name")
            except Exception as e:
                print(f"Exception is {type(e)}: {e}")     
    return customer_orders
````

````python
def calculate_total_price(customer_orders):
    total_price = 0
    print(customer_orders)
    for order in customer_orders:
        valid_input = False
        while not valid_input:
            try:
                price = int(input(f"Enter the price of {order} "))
                if price > 0:
                    total_price +=price
                    valid_input = True
                else:
                    print("Price cannot be negative. Please enter a valid price.")
            except ValueError:
                print("Invalid input. Please enter a valid price. It must be a number")
    print(total_price)
````

````python
def update_inventory(customer_orders, inventory):
    for product in customer_orders:
        if product in inventory:
            inventory[product] -= 1
    return inventory

def calculate_order_statistics(customer_orders, products):
    total_products_ordered = len(customer_orders)
    percentage_ordered = (total_products_ordered / len(products)) * 100
    return total_products_ordered, percentage_ordered

def print_order_statistics(order_statistics):
    print("Order Statistics:")
    print("Total Products Ordered:", order_statistics[0])
    print("Percentage of Unique Products Ordered:", order_statistics[1])

def print_updated_inventory(inventory):
    print("Updated Inventory:")
    for product, quantity in inventory.items():
          print(f"{product}: {quantity}")
````

````python
products = ["t-shirt", "mug", "hat", "book", "keychain"]
inventory = initialize_inventory(products)
customer_orders = get_customer_orders(inventory)
inventory = update_inventory(customer_orders, inventory)
order_statistics = calculate_order_statistics(customer_orders, products)
print_order_statistics(order_statistics)
print_updated_inventory(inventory)
calculate_total_price(customer_orders)
````
